```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib import rcParams # еӯ—дҪ“й…ҚзҪ®,ж”ҜжҢҒдёӯж–Ү
rcParams['font.family'] = 'SimHei'
rcParams['axes.unicode_minus'] = False  # и§ЈеҶіиҙҹеҸ·дёҚжҳҫзӨәзҡ„й—®йўҳ

import tensorflow as tf
```

```python
def f(x):
    return x**2
```

```python
# ========= з®ҖеҚ•зҡ„зәҝжҖ§жЁЎеһӢ =========
# зӣ®ж Ү: йў„жөӢ: fx = w * x + b (w:жқғйҮҚ , b:еҒҸзҪ®)
```

```python
# 1. еҲ¶йҖ ж ·жң¬ж•°жҚ®

# жӯЈзЎ®зҡ„жқғйҮҚе’ҢеҒҸзҪ®:
TRUE_W = 3.0
TRUE_B = 2.0
NUM_EXAMPLES = 201 # 201 ж•°жҚ®йҮҸ

# зңҹе®һеҮҪж•°
def f(x):
    return x * TRUE_W + TRUE_B

# е»әз«Ӣxж•°жҚ®еә“(ж ·жң¬,зӯүдјҡиҝҳиҰҒз»ҷy_dataеҠ дёҠеҷӘйҹі)
x_data = tf.linspace(-2,2,NUM_EXAMPLES) 
x_data = tf.cast(x_data , tf.float32) # зұ»еһӢиҪ¬жҚў

# з”ҹжҲҗеҷӘеЈ°ж•°жҚ®
noise = tf.random.normal(shape=[NUM_EXAMPLES])

# е»әз«Ӣyж•°жҚ®еә“,йңҖиҰҒеҠ дёҠеҷӘйҹі
y_data = f(x_data) + noise
```

```python
# 2. SequenceжЁЎеһӢ
from tensorflow.keras import layers
from tensorflow.keras import Input
model = tf.keras.Sequential([
    Input(shape=(1,)),    # дёҖз»ҙж•°жҚ® 
    layers.Dense(units=1) # жҜҸдёҖеұӮзҘһз»Ҹе…ғдёҺеүҚдёҖеұӮзҡ„жүҖжңүзҘһз»Ҹе…ғзӣёиҝһжҺҘ 
])
# жү“еҚ°жЁЎеһӢз»“жһ„
# model.summary()
# 
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
    loss = 'mean_squared_error'
)

model.fit(x_data , y_data , epochs=10)
print(model.variables)
```
