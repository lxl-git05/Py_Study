```python
def tensor_print(t , num=0):
    print(f"{t} , type:{type(t)}" , end='')
    print(f' row:{t.shape[0]} , columns:{t.shape[1]} , last:{t.shape[-1]}')
    if num != 0:
        print('-' * 15,end='')
        print(f' 第{num}项如上',end='')
        print('-' * 15)
    else:
        print('-'*30)
```

## 2-1 张量的基本运算
</font>

```python
# 创建张量
t1 = torch.tensor([1,2,3])

# 1. 加法 , 其实符号运算更方便: +
t2 = t1.add(10)
tensor_print(t2 , 1)

# 2. 减法(sub)
t2 -= 10
tensor_print(t2 , 2)

# 3. 乘法
t2 = t1.mul(2)
tensor_print(t2 ,3 )

# 4. 除法
t2 = t1.div(2)
tensor_print(t2 ,4)

# 5. 取反
t2 = t1.neg()
tensor_print(t2 ,5)

# 6(其实就是3). 点乘(*):对应相乘而不是矩阵的横竖相乘
t1 = torch.tensor([[1,2,3],[4,5,6]])
t2 = torch.tensor([[1,2,3],[4,5,6]])
tensor_print(t1*t2)

# 7. 矩阵相乘(横竖相乘) 还可以 t3 = t1.matmul(t2)
t2 = t2.reshape((3,2))
tensor_print(t1@t2) # 2x3 @ 3x2
```

## 2-2 张量的运算函数

+ 均值
+ 平方根
+ 求和
+ 指数计算
+ 对数计算等等

</font>

```python
# 定义张量
t1 = torch.tensor([
    [1,2,3],
    [4,5,6]
])

tensor_print(t1)

# 1. 有dim参数的函数

# 1-1 求和
print(t1.sum(dim=0))    # 按列自求和
print(t1.sum(dim=1))    # 按行自求和
print(t1.sum())         # 总体求和
print('-'*30)

# 1-2 最大值
print(t1.max(dim=0))    # 按列自求和
print(t1.max(dim=1))    # 按行自求和
print(t1.max())         # 总体求max
print('-'*30)

# 1-3 平均值
t1 = t1.type(torch.float32)
print(t1.mean(dim=0))    # 按列
print(t1.mean(dim=1))    # 按行
print(t1.mean())         # 总体
print('-'*30)
```

```python
# 2. 没有dim参数的函数

t1 = torch.tensor([
    [1,2,3],
    [4,5,6]
] , dtype=torch.float)

# 2-1 平方,多次方
print(t1.pow(2))
print(t1.pow(3))
print('-'*30)

# 2-2 平方根
print(t1.sqrt())
print('-'*30)

# 2-3 e^n次 log-> ln log2 -> log2 log10 -> log10
print(t1.log())
print('-'*30)

# 2-4 
```

## 2-4 张量的形状操作

</font>

```python
# 张量的形状操作
torch.manual_seed(24)
t1 = torch.randint(1,10,size=(2,3))
tensor_print(t1)

# 1. reshape
t2 = t1.reshape(1,6)
tensor_print(t2,1)

# 2. unsqueeze升维  squeeze降维

# 2-1 升维
# 在0维上添加1个维度
t2 = t1.unsqueeze(0)
tensor_print(t2,1)      # 1,2,3(1个2x3)
# 在1维上添加1个维度
t2 = t1.unsqueeze(1)    # 2,1,3(2个1x3)
tensor_print(t2,2)
# 在2维上添加1个维度
t2 = t1.unsqueeze(2)    # 2,3,1(2个3x1)
tensor_print(t2,3)

# 2-1 降维(删除所有维度为1的维度)
t1 = torch.randint(1,10,size=(2,1,3,1))
tensor_print(t1,1)

t2 = t1.squeeze()
tensor_print(t2,2)
```

```python
# 3. 维度交换
# 创建张量
torch.manual_seed(24)
t1 = torch.randint(1,10,size=(2,3,4))
tensor_print(t1)

# 1. transpose改变一对维度: (2,3,4) -> (3,2,4)
t2 = t1.transpose(0,1)  # 就是将第0个维度和第1个维度交换
tensor_print(t2)

# 2. permute改变多对维度: (2,3,4) -> (4,2,3) 
t3 = t1.permute(2,0,1)  # 将2,0,1换成当前顺序
tensor_print(t2)
```

```python
# 4. transpose等可能会改变张亮的存储顺序,导致前面的改变形状可能改变不了,所以使用contiguous函数重新连续化
torch.manual_seed(24)
t1 = torch.randint(1,10,size=(2,3))
tensor_print(t1)
# 判断张量是否连续
print(t1.is_contiguous())

# 4-1 通过view函数修改上述张量的形状:其实view和premute差不多,都是面向用户
t2 = t1.view(3,2)    # 从2,3 -> 3,2 , 仍然连续
tensor_print(t2)

# 通过transpose交换维度,交换之后不再连续
t3 = t1.transpose(0,1)
print(t3.is_contiguous())

# 通过contiguous函数把不连续的张量转为连续张亮
try:
    t4 = t3.view(2,3)
except Exception as e:
    print('无法对通过transpose交换维度后的t3进行reshape,原因:',e)
finally:
    print(t3.is_contiguous())

# 所以这里使用continu
t4 = t3.contiguous().view(2,3)
tensor_print(t4)
print(t4.is_contiguous())
```

</font>

```python
torch.manual_seed(24)
t1 = torch.randint(1,10,(2,3))
torch.manual_seed(21)
t2 = torch.randint(1,10,(1,3))
torch.manual_seed(20)
t3 = torch.randint(1,10,(2,5))

tensor_print(t1)
tensor_print(t2)

# 1. cat: 不改变维度数,拼接张量,除了拼接的那个维度外,其他维度必须保持一致
t4 = torch.cat([t1,t2] , dim=0) # 0维度,其实就是mxn的m,其他维度必须保持一致
tensor_print(t4,1)

t5 = torch.cat([t1,t3] , dim=1) # 0维度,其实就是mxn的n
tensor_print(t5,1)

# 2. stack: 会改变维度数,拼接张量,所有维度都必须保持一致
# 略,累了,以后学

```
