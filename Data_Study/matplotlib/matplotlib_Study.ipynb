```python
# 1-1 折线图
# 折线图库
import matplotlib.pyplot as plt 
from matplotlib import rcParams # 字体配置,支持中文
rcParams['font.family'] = 'SimHei'

# 创建图表,大小
plt.figure(figsize=(5,2))

# ========= x , y轴数据 =========
month = [str(i) + '月' for i in range(1,5)]
sales = [100 , 200 , 80 , 130]

# ================== !配置参数! ==================
# 折线图(x轴 , y轴)
plt.plot(month , sales , label = 'A',color='orange',linewidth=2,linestyle='--')
# 添加标题
plt.title("趋势" , color = 'red' , fontsize=20)
# 添加坐标轴标签
plt.xlabel('月份' , fontsize=10)
plt.ylabel('deu'  , fontsize=10)
# 添加图例
plt.legend(loc = 'upper left')
# 添加网格线
# plt.grid(axis='x')
plt.grid(True , alpha=0.1 , color='blue' , linestyle='--')
# 设置刻度字体大小
plt.xticks(rotation=0 , fontsize=12)
plt.yticks(rotation=0 , fontsize=12)
# 设置y轴范围
plt.ylim(20,240)
# 在每个数据点上显示数值
for x,y in zip(month , sales):
    plt.text( x , y+1 , str(y) , ha='center' , va='bottom' , fontsize=10)

# ================== 显示图表 ==================
plt.show()
```

```python
# 1-2 折线图
# 折线图库
import matplotlib.pyplot as plt 
from matplotlib import rcParams # 字体配置,支持中文
rcParams['font.family'] = 'SimHei'

# 创建图表,大小
plt.figure(figsize=(5,2))

# ========= x , y轴数据 =========
subjects = ['语文' , '数学' , '英语' , '物理']
score = [85 , 92 , 78 , 88]

# ================== !配置参数! ==================
# 柱状图 bar
plt.bar(subjects , score , label = '分数',color='orange' , width=0.6)
# 添加标题
plt.title("成绩分布" , color = 'red' , fontsize=20)
# 添加坐标轴标签
plt.xlabel('科目'  , fontsize=10)
plt.ylabel('分数'  , fontsize=10)
# 添加图例
plt.legend(loc = 'upper left', fontsize=5)
# 添加网格线
plt.grid(axis='y' , alpha=0.1)
# 设置刻度字体大小
plt.xticks(rotation=0 , fontsize=12)
plt.yticks(rotation=0 , fontsize=12)
# 设置y轴范围
plt.ylim(0,100)
# 在每个数据点上显示数值
for x,y in zip(subjects , score):
    plt.text( x , y-1 , str(y) , ha='center' , va='bottom' , fontsize=10)
# 自动优化排版
# plt.tight_layout()

# ================== 显示图表 ==================
plt.show()
```

```python
# 1-3 条形图:barh
# 折线图库
import matplotlib.pyplot as plt 
from matplotlib import rcParams # 字体配置,支持中文
rcParams['font.family'] = 'SimHei'

# 创建图表,大小
plt.figure(figsize=(5,2))

# ========= x , y轴数据 =========
subjects = ['语文' , '数学' , '英语' , '物理']
score    = [85 , 92 , 78 , 88]

# ================== !配置参数! ==================
# 条形图
plt.barh(subjects , score , label = '分数',color='red')
# 添加标题
plt.title("成绩分布" , color = 'red' , fontsize=20)
# # 添加坐标轴标签
plt.xlabel('科目'  , fontsize=10)
plt.ylabel('分数'  , fontsize=10)
# 添加图例
plt.legend(loc = 'upper left', fontsize=5)
# 添加网格线
plt.grid(axis='y' , alpha=0.1)
# 设置刻度字体大小
plt.xticks(rotation=0 , fontsize=12)
plt.yticks(rotation=0 , fontsize=12)
# 自动优化排版
# plt.tight_layout()

# ================== 显示图表 ==================
plt.show()
```

```python
# 1-4 饼图: pie
# 导入库
import matplotlib.pyplot as plt 
from matplotlib import rcParams # 字体配置,支持中文
rcParams['font.family'] = 'SimHei'

# 创建图表,大小
plt.figure(figsize=(8,4))

# ========= x , y轴数据 =========
things = ['学习' , '娱乐' , '运动' , '睡觉' , '其他']
times  = [6 , 4 , 1 , 8 , 5]
colors = ['red' , 'orange' , 'green' , '#66b3ff' , '#ffcc99' , '#ff4499']
explode = [0,0,0,0,0.3] # 设置突出块的位置

# ================== !配置参数! ==================
# 饼图: 
plt.pie(times ,labels=things , 
        autopct='%.1f%%' ,  # 两个% : 转义
        startangle=80 ,     # 旋转角度
        colors=colors ,     # 配色
        wedgeprops={'width':0.7},       # !!!饼图变成环形图!!!
        pctdistance=0.65,   # 百分比数据的位置
        explode=explode,    # 爆炸效果
        shadow=True         # 阴影效果
    )    
# 添加标题
plt.title("一天的时间分布" , color = 'black' , fontsize=10)
plt.text(0,0,'总计24h',ha='center',va='bottom',fontsize=10)
# 自动优化排版
plt.tight_layout()

# ================== 显示图表 ==================
plt.show()
```

```python
# 1-5 散点图
# 导入库
import matplotlib.pyplot as plt 
from matplotlib import rcParams # 字体配置,支持中文
rcParams['font.family'] = 'SimHei'

# 创建图表,大小
plt.figure(figsize=(10,5))

# ========= x , y轴数据 =========
import random
x_data = []
y_data = []
for i in range(1000):
    tmp = random.uniform(0,10)
    x_data.append(tmp)
    tmp2 = 2*tmp + random.gauss(0,2)    # 随机高斯参数
    y_data.append(tmp2)
# ================== !配置参数! ==================
# 散点图
plt.scatter(x_data , y_data,color='blue',alpha=0.5,s=6)
# 添加标题
plt.title("x变量与y变量的关系" , color = 'black' , fontsize=20)
# 添加坐标轴标签
plt.xlabel('x' , fontsize=10)
plt.ylabel('y' , fontsize=10)
# 添加图例
plt.legend(loc = 'upper left')
# 添加网格线
# plt.grid(axis='x')
plt.grid(True , alpha=0.1 , color='blue' , linestyle='--')
# 设置刻度字体大小
plt.xticks(rotation=0 , fontsize=12)
plt.yticks(rotation=0 , fontsize=12)
# 设置y轴范围
plt.ylim(0,30)
plt.plot([0,10],[0,20],color='red') # 回归线
# ================== 显示图表 ==================
plt.show()
```

```python
# 1-7 多图绘制
import matplotlib.pyplot as plt 
from matplotlib import rcParams 
rcParams['font.family'] = 'SimHei'
# 要绘图的数据
month = [str(i) + '月' for i in range(1,5)]
sales = [100 , 200 , 80 , 130]

# 创建多图(动态生成)
f1 = plt.subplot(2,2,1) # 创建子图1 , 行 列 索引
f1.plot(month , sales)

f2 = plt.subplot(2,2,2) # 创建子图2
f2.bar(month , sales)

f3 = plt.subplot(2,2,3) # 创建子图3
f3.scatter(month , sales)

f4 = plt.subplot(224) # 创建子图4,一般默认单数,所以可以缩写
f4.barh(month , sales)
```

```python
# 案例:数据可视化: 温度分析
# 1. 导入库
import pandas as pd 
import matplotlib.pyplot as plt 
from matplotlib import rcParams 
rcParams['font.family'] = 'SimHei'
rcParams['axes.unicode_minus'] = False  # 解决负号不显示的问题

# 2. 创建图表,大小
plt.figure(figsize=(10,8))

# 3. 导入数据
df = pd.read_csv('../data/weather.csv')
print(type(df['date'][0]))
df['date'] = pd.to_datetime(df['date'])

# 4-1. 绘制气温的趋势变化图
# 只保留2012年的数据,不然太多了
df = df[df['date'].dt.year == 2012]
# 绘制折线图:两条
plt.plot(df['date'] , df['temp_max'] , label = '最高气温',color='orange',linewidth=2)
plt.plot(df['date'] , df['temp_min'] , label = '最低气温',color='blue'  ,linewidth=2)
# === 其他性质 ===
# 添加标题
plt.title("2012年温度趋势" , color = 'green' , fontsize=20)
# 添加坐标轴标签
plt.xlabel('日期' , fontsize=10)
plt.ylabel('温度' , fontsize=10)
# 添加图例
plt.legend(loc = 'upper left')
# 添加网格线
plt.grid(True , alpha=0.1 , color='blue' , linestyle='--')
# 设置刻度字体大小
plt.xticks(rotation=0 , fontsize=12)
plt.yticks(rotation=0 , fontsize=12)
# 设置y轴范围
plt.ylim(-10,40)
# 绘制图表
plt.show()

# 4-2. 绘制降水量的直方图
plt.hist(df['precipitation'] , bins=5) # 区间分箱:直方图
plt.show()
```

```python
# 导入库
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from matplotlib import rcParams 
rcParams['font.family'] = 'SimHei'
rcParams['axes.unicode_minus'] = False  # 解决负号不显示的问题

# 导入数据
penguins = pd.read_csv('../data/penguins.csv')
penguins.dropna(inplace=True)   # 删除空数据
penguins.info()
```
