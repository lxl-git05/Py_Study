```python
# 1-2 Series็ๅฑๆง
s = pd.Series([32,21,54,99,3] , index=["a" , "b" ,"c" ,"d" ,"e"] , name="็คบไพ")
# print(f"s = \n{s}")
# 1. index
print(s.index)
# 2. values
print(s.values)
# 3. shape , ndim , size
print(s.shape , s.ndim , s.size)
# 4. dtype name 
print(s.dtype , s.name)
# 5. ๆพๅผ็ดขๅผ: loc + at ้ๅผ็ดขๅผ: iloc + iat
print(s.loc["a":'c'])   # ๅฑ็คบๅ็ๆไฝ
print(s.iloc[0:2])
print(s.at['a'])        # ๆพ็คบ,ไธๆฏๆๅ็
print(s.iat[0])
```

```python
# 2-1 ๅธธ็จๆนๆณ
# mode: ไผๆฐ
# quantile(q) , ๅไฝๆฐ , 0-1ไน้ด
# describe : ๅคไธชๆง่ดจๆๅฐ

# ๅธธ่งๅฝๆฐ
s = pd.Series([10,4,np.nan,None,3,4,5] , index = ["A" , "B" , "C" , "D" , "E" , "F" , "G"])
# 1. ๅ5่ก(้ป่ฎค) + ๅ2ไธช
# print(f"s.head() , s.tail(2) = {s.head()} ,\n{s.tail(2)}")
# 2. ๆฅ็ๆๆๆ่ฟฐไฟกๆฏ
s.describe()    # ็ผบๅคฑๅผNanไธ่ฎกๅฅcount
# 3. ๅ็ด ไธชๆฐ
print(s.count())
# 4. ่ทๅ็ดขๅผ
print(s.keys())
print(s.index)
# 5. ๆฅๆพ็ผบๅคฑๅผ
# print(s.isna())
# 6. ๆฅๆพๆฐๆฎๆฏๅฆๅญๅจ(ๅ่กจไธบๅฝขๅ)
print(s.isin([4,10]))
# 7. ็ป่ฎกๅญฆๆไฝ
print(s.mean() , s.var() , s.std() , s.sum() , s.min() , s.max() , s.median() , s.quantile(0.2))
print(f"ไผๆฐ:\n{s.mode()}")
# 8. ๅป้
print(s.drop_duplicates())  # ๅพๅฐpd
print(s.unique())           # ๅพๅฐlist
# 9. ๆๅบ
s.sort_index()  # ๆ็ดขๅผๆๅบ
s.sort_values() # ๆๅผๆๅบ
```
