```python
# 4-2 缺失值的处理
import pandas as pd
import numpy  as np

# 列举缺失值
s = pd.Series([1,2,np.nan,None,pd.NA])
df= pd.DataFrame([[1,pd.NA,2] , [2,3,4] , [None , 4 ,6 ]] , columns=['A' , "B" , 'C'])
# 查看缺失值
# print(s.isna())
# print(s.isnull())
print(df.isna())
print(df.isnull())
print(df.isna().sum(axis=1))    # 按照行查看缺失值的个数
print('-'*30)
# 处理缺失值
# 1. 剔除缺失值: 要注意,每一行对应的是采集的一次数据,而每一列只是数据的众多性质(列)中的一个
print(s.dropna())
print(df.dropna())          # 剔除的是一整条的记录
print(df.dropna(how="all")) # 所有的值都是缺失值才剔除这一行
print(df.dropna(thresh=3))  # 至少有n个值不是缺失值才保留
print(df.dropna(thresh=3,axis=1))  # 至少有n个值不是缺失值才保留 , 对列操作(axis = 0就对行操作)
print(df.dropna(subset=['A']))  # 只针对A列
```

```python
# 4-5 数据变形
data = {
    'ID':[1,2],
    'name':['A haha' , "B haha"],
    'Math':[90,85],
    'English':[88,92],
    'Science':[95,89],
}
df = pd.DataFrame(data)

df.T    # 转置
# 宽表转换成长表
# 解读:ID和name为自带属性不会变化,var_name为其他属性,共用一个名字列,value_name为其他属性共用的数据列
df2 = pd.melt(df,id_vars=['ID' , 'name'],var_name='科目',value_name='分数')   
df2.sort_values('name')

# 长表转换成宽表
pd.pivot(df2 , index=['ID' , 'name'] , columns='科目' , values='分数')

# 分列: expand扩展分列,必须要有!!!
# 例1 姓名空格分列
df[['first' , 'last']] = df['name'].str.split(' ' , expand=True)    # .str转为str 
# 例2 给blood_pressure分列
df = pd.read_csv('../data/sleep.csv')
df[['high' , 'low']] = df['blood_pressure'].str.split('/' , expand=True).astype('int16')
df.head(2)
```

```python
# 4-6 数据分箱 pd.<q>cut(x,bins,labels) bins:给数据分为n段区间,起始值和结束值为min和max
import pandas as pd
df = pd.read_csv('../data/employees.csv')
df1 = df.head(10)[['employee_id' , 'salary']]
# print(df1)
# 分箱(其实就是某列数据分区间)
pd.cut(df1['salary'] , bins = 3)    # 区间分隔值
df1['收入范围'] = pd.cut(df1['salary'] , bins=[0,10000,20000,30000],labels=['低','中','高'])  # 手动配置区间
pd.qcut(df1['salary'] , 3)   # 等距分隔(等人数/...)

# 例:对睡眠质量进行分箱评估
df = pd.read_csv('../data/sleep.csv')
df['睡眠质量'] = pd.cut(df.sleep_quality , bins=3 , labels=['低' , '中' , '高'])

# 对分箱的类别就是category
# str -> 类别 -> 统计
# 数值-> 分箱  -> 统计 , 都是category

# 重命名
data = {
    'name': ['A' , "B" , 'A' , 'A'] ,
    "age" : [26,25,26,30],
    'city': ['NY' , 'LA' , "NY", 'CN'],
}
df = pd.DataFrame(data)
# df.set_index('name')    # 将某一列设为索引
df = df.rename(columns={'age':"年龄"},index={0:4})   # 也是使用字典进行rename(列名或者序列号)
print(df.at[4,'年龄'])
```

```python
# 4-8 分组聚合(统计) 先分组再统计
# df.groupby('分组的字段)['聚合的字段'].聚合函数(sum , mean等)
import pandas as pd 
df = pd.read_csv('../data/employees.csv')

# 数据预处理
# id填充(也可以删去)
df['department_id'] = df.department_id.ffill()
# 员工id float转为int
df['department_id'] = df['department_id'].astype('int64')

# 数据处理(分组聚合还可以多个参数)
df.groupby('department_id').groups  # 根据department_id进行分组,同一个组的department_id相同,组内是序列号
df.groupby('department_id').get_group(90)   # 查看具体的某个分组数据
df2 = df.groupby('department_id')[['salary']].mean().round(2)  # 先根据department_id分组,再围绕salary进行同组数据处理
df2.sort_values('salary')
```

```python
# 案例1 企鹅数据分析
# 1. 导入必要的库
import pandas as pd 
# 2. 导入数据
df = pd.read_csv('../data/penguins.csv')
df.head()
# df.info()
# 3. 数据清洗
# 缺失值检查
df.dropna(inplace=True) # 直接去除缺失值
df.isna().sum()
# 4. 数据特征的构造
# 4-1 性别数据类型转换
# df['sex'] = df['sex'].map({"Male":True , "Female":False}) # 第一种:bool值
df['sex'] = df['sex'].astype('category')   # 第二种:category
# 4-2 比例
df['bill_ratio'] = df['bill_length_mm'] / df['bill_depth_mm']
# 5. 数据分析
# 体重数据分箱
labels = ['低' , '中' , '高']
df['体重'] = pd.cut(df['body_mass_g'] , bins=3 , labels=labels)
# 按照岛屿分析体重(分组聚合)
df.groupby('island')['体重'].value_counts()
```

```python
# 案例2 睡眠质量分析
# 1. 导入必要的库
import pandas as pd 
# 2. 导入数据
df = pd.read_csv('../data/sleep.csv')
# df.head()
# df.info()

# 3. 数据清洗
# 缺失值检查
df.isna().value_counts()
df.drop(columns='sleep_disorder' , inplace=True)    # 直接删除大量缺失的sleep_disorder列
# 4. 数据特征的构造
# 数据类型转换
df['gender'] = df['gender'].astype('category')
df['occupation'] = df['occupation'].astype('category')
df['bmi_category'] = df['bmi_category'].astype('category')
# 数据分列
df[['high' , 'low']] = df['blood_pressure'].str.split("/" , expand=True)
# 数据分箱:睡眠质量 + 年龄
df['睡眠质量'] = pd.cut(df['sleep_quality'] , bins = 3 , labels=['差' , '中' , '优'])
df['age'] = pd.cut(df['age'] , bins = 3 , labels=['青年' , '中年' , '老年'])
# 5. 数据统计,分析
# 查看BMI分布
df.groupby('bmi_category').agg({
    'sleep_duration' : 'mean',
    'sleep_quality' : 'mean',
    'stress_level' : 'mean',
})

# df.info()
# df.head()
```
