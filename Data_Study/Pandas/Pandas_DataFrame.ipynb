```python
# 3-1 建立DF
# 1. Series建立DF
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([2,3,4,5,6])
df = pd.DataFrame({"第1列":s1 , "第2列":s2})
# print(type(df))
# print(df)
# print(df["第1列"])

# 2. 字典创建DF
df = pd.DataFrame(
    {
        "id":[1,2,3],
        "name":["a" , "b" , "c"],
        "score":[2.3 , 3.2 , 9.0],
    }, index = ['1' , 'b' , 3] , columns=["name" , "score" , "id"] # columns可以更改列的排序
)
print(df)
```

```python
# 3-2 DF属性
# 和Series一样的
'''
index values dtypes shape ndim size columns 
loc iloc at iat T(转置)
'''
# 1. index columns
print(f"行索引:{df.index} , 列标签:{df.columns}")
# 2. values
print(df.values)
# 3. dtypes shape
print(df.dtypes , df.shape)
# 4. T转置
print(df.T)
print(f"转置前:行索引:{df.index} , 列标签:{df.columns}\n转置后:行索引:{df.T.index} , 列标签:{df.T.columns}")
# 5-1. 获取元素 loc指向行
# print(df.loc['b'])
# print(df.iloc[0])
# 5-2. 获取元素 loc指向列
# print(df.loc[:,"name"])
# print(df.iloc[:,1])
# print(df.name)    # 比较方便
# print(df['name']) # 最方便
# 5-3. 获取单个元素 at / loc 行 , 列
# print(f"df.at['1','score'] = {df.at["1",'score']}")
print(f"df.iat[0,0] = {df.iat[0,0]}")
```

```python
# 3-3 DF常用函数(其实与Series一样的)
# 1. 采样查看数据
print(df.tail(2))
print(df.head(1))
# 2. bool运算
df[(df.score>3) & (df.id < 3)]
# 3. 随机抽样
print(df.sample())
# 4. isin:查看元素是否包含在参数集合中,只要放在list里面都会去查看
print(df.isin(['c' , 2]))
# 5. isna: 查看元素是否是缺失值
# 6. sum max min var std
print(df.sum()) # str会拼接,没有统计意义,一般还是求某一列(Series)的和
print(df.score.idxmin())
print(df['id'].median())
# 7. describe
# print(df.describe())
# 8-1. count : Count non-NA cells for each column or row.
df.iat[1,2] = 1
print(df.count())   # 统计
# 8-2. value_counts(各个元素重复的次数)
print(df.id.value_counts())
# 9. replace(替代) cumsum(累积和)
# 10. 排序
# 10-1 按照索引
# print(df.sort_index())
# 10-2 按照值
print(df.sort_values(by=['id' , 'score'],ascending=[False , True])) # 填写by作为排序参考
print(df.score.sort_values())                                       # 自己选取Series进行排序函数
# 11. 求最大/小值
print(f"df.nlargest(想要的个数,columns=['想要比较的Series']) = \n{df.nlargest(2,columns=['score'])}")
print(f"df.id.nsmallest(1) = \n{df.id.nsmallest(2)}")
```

```python
# 案例1 学生成绩分析
import  numpy as np
import pandas as pd
data = {
    '姓名': ['张三' , '李四' , '王五' , '赵六' , '钱七'],
    '数学': [85 , 92 , 78 , 88 , 95],
    '英语': [90 , 88 , 85 , 92 , 80],
    '物理': [75 , 80 , 88 , 85 , 90],
}
# 创建df
score = pd.DataFrame(data=data)
# 总分和平均分
# print([score.iloc[i , 1:].sum() for i in range(4)])
score['总分'] = score[['数学' , '英语' , '物理']].sum(axis=1)
print(f"平均分:\n{[score['总分'] / 3]}")
# 数学高于90 or 英语高于85
print(score[(score['数学'] > 90) | (score['英语'] > 85)])
# 总分从高到低排序,输出前3
print(f"score.nlargest(3 , ['总分']) = {score.nlargest(3 , ['总分'])}")
```

```python
# 案例2
data = {
    '产品名称':['A' , 'B' , 'C' , 'D'],
    '单价': [100,150,200,120],
    '销量': [50,30,20,40]
}
df = pd.DataFrame(data)
# 总额 = 单价 * 销量
df['销售额'] = df['单价'] * df['销量']
print(df)
# 总额最高
# print(f"总额最高:\n{df.iloc[df['销售额'].idxmax() , :]}")
print(df.nlargest(1,columns=['销量']))
# 售额从高到低排序，并输出所有产品信息
print(df.nlargest(4,  columns=['销量']))
```
